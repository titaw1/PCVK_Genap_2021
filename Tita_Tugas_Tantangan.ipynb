```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
# Perkalian Skalar

matriksA = np.array([[1,2,3,4],
                     [2,1,6,8],
                     [3,6,7,1]])
matriksB = np.array([[2,2,8,5],
                     [1,1,9,0],
                     [3,3,5,7]])   

perkalian_biasa = matriksA*matriksB

print("Matriks A :")
print(matriksA)

print("Matriks B :")
print(matriksB)

print("Perkalian Biasa Matriks : ")
print(perkalian_biasa)
```

```python
# Perkalian Dot Matriks
matriksA = np.array([[2,1,4],
                     [1,2,8],
                     [5,9,10]])
matriksB = np.array([[2,2],
                     [1,1],
                     [3,3]]) 

perkalian_matriks1 = np.dot(matriksA,matriksB)
perkalian_matriks2 = matriksA.dot(matriksB)

print("Matriks A :")
print(matriksA)

print("Matriks B :")
print(matriksB)

print("Perkalian Matriks Cara 1 : ")
print(perkalian_matriks1)

print("Perkalian Matriks Cara 2 : ")
print(perkalian_matriks2)
```

```python
# Apply identity kernel
gambar = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/lena.jpg')

kernel1 = np.array([[1,0,0],
                     [0,1,0],
                     [0,0,1]])
 
identity = cv.filter2D(src=gambar, ddepth=-1, kernel=kernel1)

final_frame = cv.hconcat((gambar, identity))
cv2_imshow(final_frame)
```
